## Data cleaning

#### Remove rows with NaN values

```python
all_data = all_data.dropna(how='all')
all_data
```

#### Find error rows and delete them

```python
temp_df = all_data[all_data['Order Date'].str[0:2] == 'Or']
temp_df
```

## Augment data with more columns

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data
```

#### Add City and State columns

```python
# Extract city and state from address, put into new 'City' and 'State' colummns next to Address column

# One method:
# city_col = all_data['Purchase Address'].str.split(',', expand=True)[1]
# state_col = all_data['Purchase Address'].str.split(',', expand=True)[2].str.split(' ', expand=True)[1]

# Another: use .apply()
def get_city(address):
    city = address.split(',')[1].strip()
    state = address.split(',')[2].split(' ')[1].strip()
    return city + ', ' + state
    
city_col = all_data['Purchase Address'].apply(lambda x: get_city(x))

all_data.insert(6, 'City', city_col)

all_data
```

## Q4: What products are most often sold together?

```python
all_data
```

```python
# Filter and keep only rows with duplicate Order IDs (2+ items in an order)
df = all_data[all_data['Order ID'].duplicated(keep=False)]

df
```

```python
# Group Products bought together in an order into a new column 'Grouped'
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head(5)
```

```python
# Drop duplicate rows of 'Order ID' and 'Grouped' -> only one row for each Order ID
df = df[['Order ID', 'Grouped']].drop_duplicates()
df.head(5)
```

```python
from itertools import combinations
from collections import Counter

# Reference: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary
# Count the pairs of products that occur most frequent
# Create Counter object
count = Counter()
# Get each row element of column 'Grouped'
for row in df['Grouped']:
    # split 'Grouped' element by ','
    row_list = row.split(',')
    # update Counter object with the combinations found in that cell element (aka that order)
    count.update(Counter(combinations(row_list, 2)))   # To get n common products together: ...combinations(row_list, n)
    
# Convert Counter object to DataFrame
new_df = pd.DataFrame.from_dict(count, orient='index',columns = ['Times Purchased'])
new_df.sort_values('Times Purchased', ascending=False, inplace=True)

new_df
```

```python
# Clean up dataframe
new_df['Product Combinations'] = new_df.index
new_df = new_df[['Product Combinations', 'Times Purchased']]
# Convert tuples got from Counter to be strings
def tup_to_str(tup):
    return ', '.join(tup)

new_df['Product Combinations'] = new_df['Product Combinations'].apply(tup_to_str)

new_df.reset_index(drop=True, inplace=True)

new_df.head(10)
```

## Q6: What are the top 3 most sold products for each month?

```python
all_data.groupby(['Product', 'Month'])['Quantity Ordered'].sum()['iPhone']
```

```python
products_sales_by_month = all_data.groupby(['Month', 'Product'])[['Quantity Ordered', 'Sales']].sum()
products_sales_by_month
```

```python
products_sales_by_month.sort_values(by=['Month', 'Quantity Ordered'], ascending=[True, False])
```

```python
# Reuse the months list already created
# print(months)
# Create new empty DataFrame with same column names as existing by_month dataframe
empty_df = pd.DataFrame(columns=products_sales_by_month.columns)
quantity_dfs = []
sales_dfs = []

for month in months:
    # Create custom Index object for Month
    month_index = pd.Index(data=np.ones(3)*month, dtype=int, name='Month')
    # Top products by quantity sold
    # Get 3 rows which have the largest value of Quantity Ordered, reset index to create new Product column
    top_quantity_products = products_sales_by_month.loc[month].nlargest(3, 'Quantity Ordered').reset_index()
    # Set index using custom index created above to represent month
    top_quantity_products.set_index(month_index, inplace=True)
    # Append result to list for dfs concatenation 
    quantity_dfs.append(top_quantity_products)
    
    # Top products by revenue
    # Get 3 rows which have the largest value of Sales, reset index to create new Product column
    top_sales_products = products_sales_by_month.loc[month].nlargest(3, 'Sales').reset_index()
    top_sales_products.set_index(month_index, inplace=True)
    sales_dfs.append(top_sales_products)
    
top_products_quantity_month = pd.concat(quantity_dfs)
top_products_sales_month = pd.concat(sales_dfs)
```
